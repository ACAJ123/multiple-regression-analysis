```python
X= df[['Radio', 'Social Media' , 'TV_Low' , 'TV_Medium']].dropna()
X = X.astype(float)
y =  df [[ 'Sales']]
X = sm.add_constant(X)
model = sm.OLS(y,X).fit()
print(model.summary())





```

The Adj. R-squared is 0.903 and that my marketing variable explains 90.3% of my sales variance. 
All except Social Media are statistically significant with a value that is less than 0.05

```python
fitted_values = model.fittedvalues
residuals = model.resid
plt.figure(figsize =(12,5))
plt.subplot(1,2,1)
plt.scatter(fitted_values, residuals, alpha=0.5)
plt.axhline(y=0, color = 'red' , linestyle ='--')
plt.title('Residuals vs Fitted Values (Linearity & Homoscedasticity)')
plt.xlabel('Fitted values')
plt.ylabel('Residuals')
plt.subplot(1, 2, 2)
stats.probplot(residuals , dist = 'norm' , plot =plt)

plt.title('Normal Q-Q Plot')
plt.tight_layout()
plt.show()









```

Executive Conclusion: Our marketing mix optimization model successfully explains 90.33% of the variance in sales performance, proving to be an exceptionally strong framework for strategic capital planning. Statistical testing reveals that TV tiers and Radio investments are highly significant drivers of revenue growth. Conversely, Social Media spend is completely non-significant (\(p = 0.824\)), indicating that additional budget directed there yields no measurable impact on sales. Diagnostic evaluations confirm that all linear regression assumptions—linearity, homoscedasticity, and normality—are fully satisfied, validating the accuracy of these conclusions.
